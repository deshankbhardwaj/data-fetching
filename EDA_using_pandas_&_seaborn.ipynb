#What percentage of users are subscribers?

```python
num_subs = df['usertype'].loc[df['usertype'] == 'Subscriber'].count()
num_subs
```

```python
total_users = df['usertype'].count()
total_users
```

```python
print(round(num_subs/total_users*100,2),'% of total users have active annual subscription')
```

# **which bike stations are the most popular spots to start trips from?**

```python
df2 = df.groupby(['start station id']).size().reset_index(name = 'number of trips')
```

```python
df2
```

```python
temp_d = df.drop_duplicates('start station id')      #created a new df to join longitudes and latitutes to the df2 df
```

```python
df2 = pd.merge(df2,temp_d[['start station id', 'start station name','start station latitude', 'start station longitude']], how = 'left', on = ['start station id'])
```

```python
df2
```

```python
sns.scatterplot(x = 'start station id', y = 'number of trips', data = df2)
```
